### Cleaning DB

```python
unsa.columns
```

Changing column names

```python
unsa.rename(columns={'NRO':'nro','NOMBRE':'nombre','DNI':'dni','INSTITUCION ':'institucion','AREA':'area',
                     'FACULTAD':'facultad','DEPARTAMENTO':'departamento','CARGO':'cargo','AREA':'area',
                     'CORREO':'correo_65'}, inplace=True)
```

Filling emty values

```python
unsa.institucion.fillna(unsa.institucion.mode()[0], inplace=True)
unsa.area.fillna('otros', inplace=True)
unsa.departamento.fillna('AUTORIDADES PRINCIPALES', inplace=True)
unsa.cargo.fillna('DOCENTE', inplace=True)
```

Extract numbers to DNI column 

```python
unsa.dni = unsa.nombre.str.extract(r'(\d+)')
unsa.dni
```

Cleaning column nombre

```python
unsa.nombre = unsa.nombre.str.lower()
unsa.nombre
```

```python
unsa.nombre = unsa.nombre.str.replace('\d+','')
unsa.nombre = unsa.nombre.str.replace(',','')
unsa.nombre
```

#### Trying to guess and create emails

- **@unsa.edu.pe**


- NORMALLY: FIRST LETTER FIRST NAME, FIRST LAST NAME, FIRST LETTER SECOND LAST NAME (65%)

- SPECIAL CASES    
    - FIRST LETTER FIRST NAME, FIRST LAST NAME (30%)

    - FIRST LETTER SECOND NAME, FIRST LAST NAME, FIRST LETTER SECOND LAST NAME (5%)
        - luis Mauricio HUACO Zúñiga: mhuacoz@unsa.edu.pe

Removing whitespaces

```python
unsa.nombre = unsa.nombre.str.rstrip()
unsa.nombre = unsa.nombre.str.lstrip()
```

Creating emails

```python
unsa[['nombre_a','segundo_nomre','apellido']] = unsa.nombre.str.split(' ',2,expand=True)
```

```python
unsa[['apellid_1','apellid_2' ]]=unsa.apellido.str.split(' ',1,expand=True)
```

```python
unsa['first_d_name'] = unsa['nombre'].astype(str).str[0]
```

```python
unsa['first_d_last'] = unsa['apellid_2'].astype(str).str[0]
```

```python
#Concatenate correo_65%
unsa.correo_65 = unsa.first_d_name.str.cat(unsa.apellid_1, sep='')
```

```python
unsa.correo_65 = unsa.correo_65.str.cat(unsa.first_d_last, sep='')
```

```python
unsa.correo_65 = unsa.correo_65 + '@unsa.edu.pe'
```

```python
#Concatenate correo 30% 
unsa.nombre_a = unsa.nombre_a.str.cat(unsa.apellid_1,sep='')
```

```python
unsa.nombre_a = unsa.nombre_a + '@unsa.edu.pe'
```

```python
unsa.rename(columns={'nombre_a':'correo_30'},inplace=True)
```

```python
unsa.drop(columns=['segundo_nomre','apellido','apellid_1','apellid_2','first_d_name','first_d_last'], inplace=True)
```
